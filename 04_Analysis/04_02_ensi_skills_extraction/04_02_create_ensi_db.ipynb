```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore", category=DeprecationWarning) 
import requests
import pandas as pd
from pandas import json_normalize # easy JSON -> pd.DataFrame
from tqdm import tqdm_notebook
from random import randint
import time
```

```python

# This code is sending a POST request to the specified URL with the client ID and client secret in the body of the request. 
# The headers variable contains a content-type header set to application/x-www-form-urlencoded, which tells the server that the body of the request is URL encoded. 
# The server will respond with some text, which is printed out by the code.


url = "https://auth.emsicloud.com/connect/token"

payload = "YOUR CLIENT ID AND CLIENT SECRET HERE"
headers = {'content-type': 'application/x-www-form-urlencoded'}

response = requests.request("POST", url, data=payload, headers=headers)

print(response.text)
```

```python
access_token = "YOUR API TOKEN HERE"
```

```python

# This code defines a function to extract a list of skills from an external API. It makes a GET request to the API endpoint, passing in the access token as a header, and then parses the response into a JSON object. 
# Finally, it returns a DataFrame containing the list of skills.

def extract_skills_list():
  all_skills_endpoint = "https://emsiservices.com/skills/versions/latest/skills" # List of all skills endpoint
  auth = "Authorization: Bearer " + access_token # Auth string including access token from above
  headers = {'authorization': auth} # headers
  response = requests.request("GET", all_skills_endpoint, headers=headers) # response
  response = response.json()['data'] # the data

  all_skills_df = pd.DataFrame(json_normalize(response)); # Where response is a JSON object drilled down to the level of 'data' key
  return all_skills_df

extract_skills_list()
```

```python
# The function extract_skills_list() returns a list of all the skills present in the dataset. 
# The head() function returns the first 5 elements of the list.

all_skills = extract_skills_list() 
all_skills.head(5)
```

```python
# The code below will create a CSV file called 'all_skills.csv' and will populate it with data from the 'all_skills' variable.
all_skills.to_csv('all_skills.csv')
```

```python
# This code reads in a dataset from a CSV file, and then displays the first five rows of data from that dataset.

df = pd.read_excel("Master/04_Analysis/04_01_custom_analysis/04_01_02_tableau_data/01_tableau_base_data_geocoded.xlsx")
df.head(5)
```

```python
# The code uses the query method to select only the rows in the dataframe that have a role of "data analyst". It then uses the head method to display the first 5 rows of data.

df_da = df.query('role == "data analyst"')
df_da.head(5)
```

```python
# The code is creating a list of job descriptions from a dataframe, and then subsetting that list to the first two items.

job_descriptions_list = df_da['jobDescription'].tolist()
job_descriptions_list = job_descriptions_list[:2]
job_descriptions_list
```

```python
# This code uses the tqdm_notebook library to show a progress bar while it loops through a list of job descriptions, sending each one to an API endpoint to get a list of skills. It then appends the skills to a list.

list_of_skills = []

for job_description in tqdm_notebook(job_descriptions_list):
  text = job_description
  skills_from_endpoint = "https://emsiservices.com/skills/versions/latest/extract"
  confidence_interval = '1'
  headers = {'authorization': "Bearer " + access_token,'content-type': "application/json"}
  payload = "{ \"text\": \"... " + text + " ...\", \"confidenceThreshold\": " + confidence_interval + " }"
  response = requests.request("POST", skills_from_endpoint, data=payload.encode('utf-8'), headers=headers)
  skills_found = json_normalize(response.json())
  skills_found_df = pd.DataFrame(skills_found)
  list_of_skills.append(skills_found_df)

  time.sleep(randint(1,3))
```

```python
# The code reads in a CSV file of all skills, prints the dimensions of the dataframe, and displays the first 5 rows.

all_skills = pd.read_csv('Master/04_Analysis/04_ensi_skills_extraction/04_all_ensi_skills_ids.csv')
print(all_skills.shape)
all_skills.head(5)
```

```python
# The code is converting the id column of the all_skills dataframe into a list.
all_skills_list = all_skills['id'].tolist()
```

```python
# The code is creating a list of URLs. It is using the tqdm_notebook function to display a progress bar for the loop. It is appending each URL to the urls list. Finally, it is printing the length of the urls list.


urls = []

for skill in tqdm_notebook(all_skills_list):
    url = "https://emsiservices.com/skills/versions/latest/skills/" + str(skill)
    urls.append(url)
```

```python
print(len(urls))
```

```python
# token is only valid for 1 hour so we need to refresh it every hour. We can only make 5 requests per second. 5000 requests per hour is the limit.

# This code is splitting a list of URLs into seven different lists, each containing 5000 URLs.

urls_1 = urls[:5000]
urls_2 = urls[5000:10000]
urls_3 = urls[10000:15000]
urls_4 = urls[15000:20000]
urls_5 = urls[20000:25000]
urls_6 = urls[25000:30000]
urls_7 = urls[30000:34268]
```

```python
# This code loops through a list of urls, sending a GET request to each one with an authorization header. It then takes the response data, transforms it into a pandas DataFrame, and appends it to a list. Finally, it sleeps for 1/5 of a second.
# Based on skill_id we will get additional information like type of skill, description, category, subcategory etc.


data_skills = []
for url in tqdm_notebook(urls_1):
    headers = {'authorization': "Bearer " + access_token,'content-type': "application/json"}
    response = requests.request("GET", url, headers=headers)
    skills_found = pd.DataFrame(json_normalize(response.json()['data']))
    skills_found = response.json()['data']
    skills_found_for_df = pd.io.json.json_normalize(skills_found)
    data_skills.append(skills_found_for_df)
    time.sleep(1/5)
```

```python
# This code reads in seven different CSV files, all of which presumably contain data on extracted skills. Each CSV file is assigned to a different variable, so that the data from each file can be accessed separately.


ensi_1 = pd.read_csv('ensi_extracted_skills_1.csv')
ensi_2 = pd.read_csv('ensi_extracted_skills_2.csv')
ensi_3 = pd.read_csv('ensi_extracted_skills_3.csv')
ensi_4 = pd.read_csv('ensi_extracted_skills_4.csv')
ensi_5 = pd.read_csv('ensi_extracted_skills_5.csv')
ensi_6 = pd.read_csv('ensi_extracted_skills_6.csv')
ensi_7 = pd.read_csv('ensi_extracted_skills_7.csv')
```

```python
# The code is creating a database of ensi skills by concatenating several smaller databases together. The final database is then exported to a CSV file.

ensi_skills_db = pd.concat([ensi_1,ensi_2,ensi_3,ensi_4,ensi_5,ensi_6,ensi_7])
ensi_skills_db.to_csv('Master/04_02_ensi_skills_extraction/04_ensi_skills_db.csv',index=False)
```
