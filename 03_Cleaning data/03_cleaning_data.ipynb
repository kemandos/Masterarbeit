```python

# This code is importing the warnings, pandas, and numpy modules. The code is also setting the warnings module to ignore FutureWarning. The code is also setting the chained_assignment option to None.


import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import numpy as np
import re
from collections import Counter
```

```python
# This code loads a CSV file into a Pandas DataFrame, then prints the number of rows and columns in the DataFrame.


df = pd.read_csv('Master/03_Cleaning data/03_data_for_cleaning.csv',encoding='utf-8')
print(df.shape)
```

```python

# This code creates a function called 'remove_empty_rows' which takes in a dataframe as an argument. 
# The function then drops all rows from the dataframe which have empty values in the columns 'jobTitle', 
# 'jobDescription', and 'Location'. The function returns the resulting dataframe. 

# The code then creates a new dataframe called 'df' by calling the 'remove_empty_rows' function on the 
# original dataframe. Finally, the code displays the first 10 rows of the new dataframe.



def remove_emtpy_rows(dataframe):
    dataframe = dataframe.dropna(subset=['jobTitle','jobDescription','Location'],how='any')
    return dataframe
df = remove_emtpy_rows(df)
df.head(10)
```

```python
# This code is designed to adjust a given dataframe in a few ways: 

# 1) Convert all dataframe values to lowercase
# 2) Drop any rows where the jobDescription column is equal to 'empty'
# 3) return the resulting dataframe
# 4) check the shape of the resulting dataframe

def general_adjustment(dataframe):

    dataframe =  dataframe.apply(lambda x: x.astype(str).str.lower())
    dataframe.drop(dataframe.loc[dataframe['jobDescription']=='empty'].index, inplace=True)

    return dataframe
    
df = general_adjustment(df)
df.shape
```

```python
# The code is creating a function that replaces all instances of '/' in the postDate column with '-', and then returns the dataframe. 
# The df variable is then assigned the result of running the cleaning_date function on the df dataframe. The final df.head(10) command just displays the first 10 rows of the df dataframe.

def cleaning_date(dataframe):
    dataframe['postDate'] = dataframe["postDate"].str.replace('/','-')
    return dataframe
df = cleaning_date(df)
df.head(10)
```

```python
# This code is designed to clean up the job descriptions in a dataframe.

# The first line defines a function that takes in a dataframe as an input.

# The second line replaces all characters in the job descriptions that are not letters, numbers, or certain special characters with a space.

# The third line replaces all multiple spaces with a single space.

# The fourth line replaces all blank spaces with NaN values.

# The fifth line strips leading and trailing spaces from the job descriptions.

# The sixth line replaces all single characters with a space.

# The seventh line replaces certain strings of text with a space.

# The eighth line returns the dataframe.

    
def cleanning_description(dataframe):
    dataframe['jobDescription'] = dataframe['jobDescription'].apply(lambda x: re.sub("[^A-Za-z0-9\u00e4\u00f6\u00fc]", " ", str(x)))
    dataframe["jobDescription"] = dataframe["jobDescription"].replace(r'\s+', ' ', regex=True) # replace multiple spaces with one space
    dataframe["jobDescription"] = dataframe["jobDescription"].replace(r'^\s*$', np.NaN, regex=True) #
    dataframe["jobDescription"] = dataframe["jobDescription"].str.strip() # remove leading and trailing spaces
    dataframe["jobDescription"] = dataframe["jobDescription"].str.replace(r"\b\w\b","").str.replace(r"\s+", " ") # remove single characters
    dataframe["jobDescription"] = dataframe["jobDescription"].str.replace('job description employment type full time', '')
    dataframe["jobDescription"] = dataframe["jobDescription"].str.replace('full job description', '')
    dataframe["jobDescription"] = dataframe["jobDescription"].str.replace('job description summary', '')
    return dataframe
df = cleanning_description(df)
df.head(10)
```

```python
# This code removes empty rows from a dataframe.

# First, the function defines a function to remove empty rows from a dataframe.
# This function takes in a dataframe as a parameter.
# Next, the function sets the dataframe equal to itself, filtered to only include rows where the 'jobDescription' column is not null.
# Finally, the function returns the dataframe.

# Next, the code sets the variable 'df' equal to the output of the 'remove_emtpy_rows' function, using the 'df' dataframe as input.
# Finally, the code displays the first 10 rows of the 'df' dataframe.


def remove_emtpy_rows(dataframe):
    dataframe = dataframe[dataframe['jobDescription'].notna()]
    return dataframe
df = remove_emtpy_rows(df)
df.head(10)
```

```python
# The code imports the "re" library in order to use regular expressions to replace certain characters in the "Company" column of the "df" dataframe.
#
# It then replaces all instances of brackets with nothing, i.e. removes them, using the "str.replace" method.
#
# It then replaces all instances of the characters "–", "-", "+" and "." with spaces, using the "str.replace" method.
#
# It then replaces all instances of multiple spaces with a single space, using the "str.replace" method and a regular expression.
#
# Finally, it returns the cleaned dataframe.

def clean_company_name(dataframe):
    # dataframe['jobDescription'] = dataframe['jobDescription'].apply(lambda x: re.sub("[^A-Za-z0-9\u00e4\u00f6\u00fc]", " ", str(x)))
    dataframe["Company"] = dataframe['Company'].str.replace(r"\(.*\)","") # remove all brackets
    dataframe["Company"] = dataframe['Company'].str.replace("–","").str.replace("-"," ").str.replace("+"," ").str.replace(".","").str.replace("|","")
    dataframe["Company"] = dataframe["Company"].replace(r'\s+', ' ', regex=True)
    return dataframe
df = clean_company_name(df)
df.head(10)
```

```python
# The code defines a function called cleaning_title, which takes a dataframe as an input.

# The function cleans up the 'jobTitle' column in the dataframe by:
# - removing all non-letter characters (including umlauts)
# - replacing multiple spaces with single spaces
# - removing leading and trailing spaces
# - converting all letters to lowercase
# - removing all single letter words
# - removing the words 'b2b', 'vollzeit', and 'wir suchen 1x'
# - removing all numbers

# The function then returns the cleaned dataframe.

# The code then calls the function on the df dataframe, and displays the first 10 rows of the cleaned dataframe.

def cleaning_title(dataframe):
    dataframe['jobTitle'] = dataframe['jobTitle'].apply(lambda x: re.sub("[^a-zA-Z&\u00e4\u00f6\u00fc]", " ", str(x)))
    dataframe["jobTitle"] = dataframe["jobTitle"].str.replace('\s+', ' ') # replace multiple spaces with single space
    dataframe["jobTitle"] = dataframe["jobTitle"].str.strip() # remove leading and trailing spaces
    dataframe["jobTitle"] = dataframe["jobTitle"].str.lower() # convert to lowercase
    dataframe["jobTitle"] = dataframe["jobTitle"].str.replace(r"\b\w\b","").str.replace(r"\s+", " ") # remove all single letter words
    dataframe["jobTitle"] = dataframe["jobTitle"].str.replace('b2b','').str.replace('vollzeit','').str.replace('wir suchen 1x','')
    dataframe['jobTitle'] = dataframe['jobTitle'].str.replace(r'\d+', '')
    
    return dataframe
df = cleaning_title(df)
df.head(10)
```

```python
#This code strips out the location data from a dataframe, leaving only the city. 
#It replaces all digits with nothing, then replaces commas and dashes with space. 
#It then removes all non-alphanumeric characters, leading spaces, and double spaces. 
#It then replaces "home" and "home office" with "remote", and "bundesweit" and "deutschland" with "remote". 
#Finally, it keeps only the first word of the location.

def clean_location(dataframe):
    dataframe['Location'] = dataframe['Location'].str.replace('\d+', '')
    dataframe['Location'] = dataframe['Location'].str.replace(",", "DELETE" ).str.replace("-", "DELIMITER" )
    dataframe["Location"] = dataframe["Location"].map(lambda x: re.sub(r"\W+", " ", x))# remove all non-alphanumeric characters
    dataframe['Location'] = dataframe['Location'].map(lambda x: re.sub(r"^\s+", "", x)) # remove all leading spaces 
    dataframe['Location'] = dataframe['Location'].str.replace("  ", " " ) # remove all double spaces
    dataframe['Location'] = dataframe['Location'].str.replace("DELETE", " " ).str.replace("DELIMITER", "-" )
    dataframe['Location'] = dataframe['Location'].str.replace("home", "remote") # r
    dataframe['Location'] = dataframe['Location'].str.replace("home office", "remote")
    dataframe['Location'] = dataframe['Location'].str.replace("bundesweit", "remote")
    dataframe['Location'] = dataframe['Location'].str.replace("deutschland", "remote")
    dataframe['Location'] = dataframe['Location'].str.split(" ").str[0] # get the first word
    return dataframe
df = clean_location(df)
df.head(10)
```

```python
# The code creates two new columns in the dataframe - 'CompanyList' and 'CompanyCounter'.
# 'CompanyList' contains a list of all the companies in the dataframe, with each company name converted to lowercase.
# 'CompanyCounter' contains a Counter object for each company in the dataframe, which counts the number of words in each company name.
# The code then drops the 'CompanyList' and 'CompanyCounter' columns, and displays the first 10 rows of the dataframe.

df['CompanyList'] = df['Company'].str.lower().str.split()
df['CompanyCounter'] = df['CompanyList'].apply(Counter)
df['CompanyWords'] = df['CompanyList'].apply(len)
df = df.drop(['CompanyList', 'CompanyCounter'], axis=1)
df.head(10)
```

```python
# This code is creating a new column in the dataframe called 'CompanyGroup'. For each row in the dataframe, it is checking to see if the number of words in the 'Company' column is greater than 3. 
# If it is, it is taking the first word of the 'Company' column and putting it in the 'CompanyGroup' column. If the number of words in the 'Company' column is not greater than 3, it is just putting the entire 'Company' column into the 'CompanyGroup' column.


df['CompanyGroup'] = np.where(df['CompanyWords'] > 3, df['Company'].str.split(" ").str[0], df['Company'])
df
```

```python

# The first two lines of code remove the & and - symbols from the CompanyGroup column. The third line removes the CompanyWords column. 
# The fourth line displays all the unique values in the jobDescription column. The fifth line resets the index. The sixth line saves the dataframe as a CSV file. 
# The seventh line displays the shape of the dataframe.

df['CompanyGroup'] = df['CompanyGroup'].str.rstrip('&')
df['CompanyGroup'] = df['CompanyGroup'].str.rstrip('-')
df.drop(columns=['CompanyWords'], inplace=True)
df['jobDescription'].unique()
df.reset_index(drop=True)
df.to_csv('03_01_cleaned_dataset.csv', index=False,encoding='utf-8')
print(df.shape)
```
