```python
import glob
import pandas as pd   
```

```python
# This code is written in Python.

# The code defines a function called 'extracting_job_postings'. This function takes one input parameter called 'input'.

# The code then imports the 'glob' and 'pandas' libraries.

# The code then creates a new dataframe called 'df_web1' by concatenating (combining) all of the CSV files in the '02_website_1' folder. The code also renames the 'jobTitel' column to 'jobTitle'.

# The code then creates a new dataframe called 'df_web2' by concatenating all of the CSV files in the '02_website_2' folder.

# The code then creates a new dataframe called 'df_web3' by concatenating all of the CSV files in the '02_website_3' folder.

# The code then creates a new dataframe called 'df_web4' by concatenating all of the CSV files in the '02_website_4' folder. The code also renames the 'posDate' column to 'postDate'.

# The code then creates a new dataframe called 'union_df' by concatenating all of the previous dataframes (df_web1, df_web2, df_web3, df_web4).

# The code then saves the 'union_df' dataframe as a CSV file called '02_union_data.csv' in the '02_union_data' folder.

# The code then saves the 'union_df' dataframe as a CSV file called '03_data_for_cleaning.csv' in the '03_cleaning data' folder.
```

```python
df_web1 = pd.concat(map(pd.read_csv, glob.glob('Master/02_Union different sources/02_website_1/*.csv')))
df_web1.rename(columns={'jobTitel': 'jobTitle'}, inplace=True)
print(df_web1.shape)
df_web1.head(5)
```

```python
df_web2 = pd.concat(map(pd.read_csv, glob.glob('Master/02_Union different sources/02_website_2/*.csv')))
print(df_web2.shape)
df_web2.head(5)
```

```python
df_web3 = pd.concat(map(pd.read_csv, glob.glob('C:/Users/Aleksej Aikov/Desktop/Enablement/Master/02_Union different sources/02_website_3/*.csv')))
print(df_web3.shape)
df_web3.head(5)
```

```python
df_web4 = pd.concat(map(pd.read_csv, glob.glob('Master/02_Union different sources/02_website_4/*.csv')))
df_web4.rename(columns={'posDate': 'postDate'}, inplace=True)
print(df_web4.shape)
df_web4.head(5)
```

```python
union_df = pd.concat([df_web1, df_web2, df_web3, df_web4])
print(union_df.shape)
union_df.head(20)
```

```python
union_df.to_csv('Master/02_Union different sources/02_union_data/02_union_data.csv', index=False)
union_df.to_csv('Enablement/Master/03_Cleaning data/03_data_for_cleaning.csv', index=False)
```
